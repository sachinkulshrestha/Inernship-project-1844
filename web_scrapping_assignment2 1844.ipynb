```python
driver=webdriver.Chrome(path)
driver.get('http://www.ambitionbox.com/')
```

```python
driver.find_element_by_xpath('/html/body/div/div/div/div[1]/header/nav/ul/li[3]/span').click()
driver.find_element_by_xpath('/html/body/div/div/div/div[1]/header/nav/ul/li[3]/div/ul/li[1]/div').click()
```

```python
driver.find_element_by_xpath('/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/input').send_keys('Data Scientist')
```

```python
driver.find_element_by_xpath("/html/body/div/div/div/main/section[1]/div[2]/div[1]/span/div/div/div[1]/div/div/p").click()
```

```python
company_name = driver.find_elements_by_xpath("//div[@class='company-info']/a")
companies=[]
for i in company_name[0:10]:
    companies.append(i.text) 
    
salary = driver.find_elements_by_xpath("//div[@class='sbold-list-header']/span")
salary_record=[]
for i in salary[0:10]:
    salary_record.append(i.text)


Average = driver.find_elements_by_xpath("//p[@class='averageCtc']")
avg_salary=[]
for i in Average[0:10]:
    avg_salary.append(i.text)
    
min_salary = driver.find_elements_by_xpath("//div[@class='salary-values']/div[1]")
min_salary_range=[]
for i in min_salary[0:10]:
    min_salary_range.append(i.text)
    
max_salary = driver.find_elements_by_xpath("//div[@class='salary-values']/div[2]")
max_salary_range=[]
for i in max_salary[0:10]:
    max_salary_range.append(i.text)
    
experience = driver.find_elements_by_xpath("//div[@class='sbold-list-header']")
experience_required=[]
for i in experience[0:10]:
    experience_required.append(i.text.split('\n')[-1].split('(')[0].strip(' '))
```

```python
df = pd.DataFrame(list(zip(companies,salary_record,avg_salary,min_salary_range,max_salary_range,experience_required)),columns=['companies','salary_record','avg_salary','min_salary_range','max_salary_range','experience_required'])
df
```

# Q4. list of respected former presidents of India

```python
page3=requests.get('https://presidentofindia.nic.in/former-presidents.htm')
page3
```

```python
content3=BeautifulSoup(page3.content)
content3
```

```python
poi=content3.find_all('div',class_=('presidentListing'))
pname=[]
for i in poi:
    for n in i.find_all('h3'):
        pname.append(n.text)
pname
```

```python
tof=content3.find_all('div',class_=('presidentListing'))
toof=[]
for i in tof:
    for t in i.find_all('p'):
        toof.append(t.text.replace('Term of Office:',''))
toof
        
    
```

```python
TOF=toof[0:8:2]+toof[8:18]
```

```python
df=pd.DataFrame(list(zip(pname,TOF)),columns=['Former Presidents','Term of Office'])
df
```

# Q7) Write a python program to scrape mentioned news details from https://www.cnbc.com/world/?region=world

```python
page10=requests.get('https://www.cnbc.com/world/?region=world')
content10=BeautifulSoup(page10.content)
content10
```

```python
Header = [] 
for header in content10.find_all(["h1","h2","h3","h4","h5","h6",]):
    Header.append(header.name+" "+header.text.strip())
Header
```

Time

```python
Time=content10.find_all('span',class_='LatestNews-wrapper')
time=[]
for i in Time:
    time.append(i.text)
time
```

Headline

```python
head=content10.find_all('a',class_='LatestNews-headline')
headline=[]
for i in head:
    headline.append(i.text)
headline
```

Link

```python
link=content10.find_all('a',class_='LatestNews-headline')
links=[]
for i in link:
    links.append(i['href'])
links
```

```python
df=pd.DataFrame(list(zip(headline,links,time)),columns=['Headline','Link','Time',])
df
```

# Q9 details from dineout.co.in 

Restaurant name

```python
page12=requests.get('https://www.dineout.co.in/delhi-restaurants/buffet-special')
content12=BeautifulSoup(page12.content)
content12
```

 Restaurant name

```python
name=content12.find_all('a',class_='restnt-name ellipsis')
names=[]
for i in name:
    names.append(i.text)
names
```

Cuisine

```python
Cuisine=content12.find_all('span',class_='double-line-ellipsis')
Cuisines=[]
for i in Cuisine:
    Cuisines.append(i.text.split('|')[1])
Cuisines
```

Location 

```python
Location=content12.find_all('div',class_='restnt-loc ellipsis')
Locations=[]
for i in Location:
    Locations.append(i.text)
Locations
```

Ratings

```python
Ratings=content12.find_all('div',class_='restnt-rating rating-4')
Rating=[]
for i in Ratings:
    Rating.append(i.text)
Rating
```

Image URL


```python
URL=content12.find_all('img',class_='no-img')
url=[]
for i in URL:
    url.append(i['data-src'])
url
```

```python
df=pd.DataFrame(list(zip(names,Cuisines,Locations,Rating,url)),columns=['Restaurant name','Cuisine','Location','Ratings','Image URL'])
df
```
